```python
import pandas as pd
```

```python
# !! config file name before execute
input_file = "1443_glidescore.csv"
output_file = "sort_1443_glidescore.csv"
```

```python
df = pd.read_csv(input_file)

result = df.groupby('title', as_index=False).apply(
    lambda x: x.loc[x['r_i_glide_gscore'].idxmin()]
)
result = result.sort_values('r_i_glide_gscore')
result[['SMILES', 'title', 'r_i_glide_gscore']].copy().to_csv(output_file, index=False)
```

```python
res = pd.read_csv(output_file)
top10_mol = res[res['r_i_glide_gscore']<-5.85].copy()
top10_mol.to_csv("sort_1443_glidescore_top10.csv", index=False)
```
